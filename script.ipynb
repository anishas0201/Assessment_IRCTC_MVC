# [dbo].[ImportExcelData]

```sql
/****** Object:  StoredProcedure [dbo].[ImportExcelData]    Script Date: 12/13/2023 12:18:05 PM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[ImportExcelData]
    @FilePath NVARCHAR(255)
AS
BEGIN
    -- Check if the table exists, and if not, create it
    IF NOT EXISTS (SELECT 1 FROM INFORMATION_SCHEMA.TABLES WHERE TABLE_NAME = 'tbl_ExcelSheet')
    BEGIN
        CREATE TABLE tbl_ExcelSheet
        (
            [S.No] INT,
            [First Name] NVARCHAR(50),
            [Last Name] NVARCHAR(50),
            [Gender] NVARCHAR(50),
            [Country] NVARCHAR(50),
            [Age] INT,
            [BirthDay] DATE
            -- Add more columns as needed
        );
    END;

    -- Insert data from the Excel file into the table
    DECLARE @SqlQuery NVARCHAR(MAX);
    SET @SqlQuery = 'INSERT INTO tbl_ExcelSheet ([S.No], [First Name], [Last Name], [Gender], [Country], [Age], [BirthDay]) ' +
                    'SELECT [S.No], [First Name], [Last Name], [Gender], [Country], [Age], [BirthDay] ' +
                    'FROM OPENROWSET(''Microsoft.ACE.OLEDB.12.0'', ''Excel 12.0;Database=' + @FilePath + ';HDR=YES;'', ''SELECT * FROM [Sheet1$]'')';

    EXEC sp_executesql @SqlQuery;

    -- Optional: Drop the table if you want to clear it after importing
    -- DROP TABLE tbl_ExcelSheet;
END
GO
```

```sql
/****** Object:  StoredProcedure [dbo].[sp_get_CollegeDetailsbyId]    Script Date: 12/13/2023 12:18:05 PM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[sp_get_CollegeDetailsbyId]
    @AdmissionId INT
AS
BEGIN
    SELECT
        ad.AdmissionId,
        ad.CollegeName,
        S.FirstName AS StudentName,
        D.BranchName,
        R.RoleDesignation,
        ad.AdmissionDate,
        ad.CollegeFees,
        ad.IsPaymentDone,
        ad.IsActive
    FROM
        tbl_AdmissionFormData ad
    INNER JOIN
        tbl_DepartmentMaster D ON D.BranchId = ad.BranchName
    INNER JOIN
        tbl_RoleMaster R ON R.RoleId = ad.RoleDesignation
    INNER JOIN
        tbl_StudentMaster S ON S.StudentId = ad.StudentName
    WHERE
        ad.AdmissionId = @AdmissionId;
END;
GO
```

# [dbo].[sp_SearchByDate]

```sql
/****** Object:  StoredProcedure [dbo].[sp_SearchByDate]    Script Date: 12/13/2023 12:18:05 PM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[sp_SearchByDate]
    @FromDate DATE,
    @ToDate DATE
AS
BEGIN
    SELECT
        ad.AdmissionId,
        ad.CollegeName,
        S.FirstName, 
        D.BranchName,
        R.RoleDesignation,
        ad.AdmissionDate,
        ad.CollegeFees,
        ad.IsPaymentDone,
        ad.IsActive
    FROM
        tbl_AdmissionFormData ad
    INNER JOIN
        tbl_DepartmentMaster D ON D.BranchId = ad.BranchName
    INNER JOIN
        tbl_RoleMaster R ON R.RoleId = ad.RoleDesignation
    INNER JOIN
        tbl_StudentMaster S ON S.StudentId = ad.StudentName
    WHERE
        ad.AdmissionDate BETWEEN @FromDate AND @ToDate;
END;
GO
```

# [dbo].[sp_validate_IRCTCLogin]

```sql
/****** Object:  StoredProcedure [dbo].[sp_validate_IRCTCLogin]    Script Date: 12/13/2023 12:18:05 PM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[sp_validate_IRCTCLogin]
 @fullName nvarchar(50),
 @emailId nvarchar(50),
    @decryptedPassword NVARCHAR(50)
AS
BEGIN
    IF EXISTS (
        SELECT 1
        FROM tbl_IRCTC_Registration
        WHERE fullName = @fullName
        AND emailId = @emailId
        AND decryptedPassword = @decryptedPassword
    )
    BEGIN
        SELECT
            su.fullName,
            su.emailId
           
        FROM tbl_IRCTC_Registration su
        WHERE su.fullName = @fullName
        AND su.emailId = @emailId
        AND su.decryptedPassword = @decryptedPassword;
    END
END;
GO
```

# [dbo].[sp_validate_LoginDetails]

```sql
/****** Object:  StoredProcedure [dbo].[sp_validate_LoginDetails]    Script Date: 12/13/2023 12:18:05 PM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[sp_validate_LoginDetails]
    @UserId INT,
    @Name NVARCHAR(50), -- Adjust the length based on your actual requirements
    @decryptedPassword NVARCHAR(50)
AS
BEGIN
    IF EXISTS (
        SELECT 1
        FROM tbl_SignUp
        WHERE UserId = @UserId
        AND Name = @Name
        AND decryptedPassword = @decryptedPassword
    )
    BEGIN
        SELECT
            su.Id,
            su.Name,
            su.UserId
        FROM tbl_SignUp su
        WHERE su.UserId = @UserId
            AND su.Name = @Name
            AND su.decryptedPassword = @decryptedPassword;
    END
END;

GO
```
